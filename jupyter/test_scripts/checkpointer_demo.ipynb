Demonstrates functionality of the checkpointer function in a recommended saving loop.
The saving loop is designed to manually dump the large data every 'n' iterations which helps clear space and combats system-wide crashes. 
The checkpointer is designed to automatically run a function on batches within a sequence, protecting against the risk of dataloss from connectivity errors or other exceptions by dumping data upon an exception being raised.

```python
from utils import checkpointer, get_subgroup
import random
import json
import itertools

my_array = [x for x in range(1000)]

batch_size=10
def sum_crash(seq):
    rand = random.random()
    if rand > 0.8: #Crash
        print("ERROR")
        raise ValueError
    else:
        return [sum(seq)]

save_file = 'all_sums_save.json'

# Initialise file
with open(save_file, 'w') as f:
    f.write('[')
# Perform function
for subgroup in get_subgroup(my_array, 100):
    # print(subgroup)
    subsum_list = checkpointer(sum_crash, subgroup, batch_size,
                        dump_file='sums_dump.json', checkpoint=0)
    with open(save_file, 'a') as f:
        json.dump(subsum_list, f)
        f.write(',\n')

with open(save_file,'a') as f:
    f.write('[]\n]')


subsums_list = []
with open(save_file) as f:
    subsums_list = json.load(f)
    # for json_object in f:
    #     subsums = json.loads(json_object)
    #     subsums_list.extend(subsums)


print(list(itertools.chain(*subsums_list)))

```
